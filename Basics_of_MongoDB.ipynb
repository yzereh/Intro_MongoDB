<a id="mongod_ID"></a>
#### <p style="color:rgb(119,136,153);">MONGOD</p>

- <font style="color:rgb(119,136,153);">The mongod process is the **core** component of **MongoDB**</font>
- <font style="color:rgb(119,136,153);">It is **responsible** for **running** the **database server**</font>
- <font style="color:rgb(119,136,153);">It handles data **storage**, management, and access</font>
- <font style="color:rgb(119,136,153);">It can run as a **standalone** server or as part of a **replica set** or **sharded cluster**</font>
- <font style="color:rgb(119,136,153);">mongod stores **data** in files on **disk**</font>


<a id="shells_ID"></a>
#### <p style="color:rgb(119,136,153);">MONGO and MONGOSH</p>

- <font style="color:rgb(119,136,153);">Both are **shells**</font>
- <font style="color:rgb(119,136,153);"> Javascript and Node.js **interfaces** that you can use to **interact** with the MongoDB **server** (mongod)</font>
- <font style="color:rgb(119,136,153);">They can be used to test **queries** and **operations** directly with your **database**</font>
- <p style="color:rgb(119,136,153);"> Mongosh is available as a <b>standalone package</b> in the <a href=https://www.mongodb.com/try/download/shell?jmp=docs style="color:rgb(216,118,0)">MongoDB dowload center</a></p>
- <font style="color:rgb(119,136,153);">mongosh can **automate** tasks, perform **batch operations**, or **schedule jobs**</font> 

<a id="mongos_ID"></a>
#### <p style="color:rgb(119,136,153);">MONGOS</p>

- <font style="color:rgb(119,136,153);">It serves as a **query** **router** in a **sharded** cluster</font>
- <font style="color:rgb(119,136,153);">It is an **intermediary** between **client** applications and the individual **mongod** instances</font>

<a id="distributed_ID"></a>
#### <p style="color:rgb(119,136,153);">DISTRIBUTED COMPUTING</p>

- <font style="color:rgb(119,136,153);">Distributed computing in MongoDB is based on the concept of **Sharding**</font>
- <font style="color:rgb(119,136,153);">Sharding is a **method** for **distributing** data across multiple **machines**</font>
